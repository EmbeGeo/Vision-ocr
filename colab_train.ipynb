### 3. 데이터셋 업로드 및 압축 해제
로컬에서 생성된 `dataset.zip` 파일을 코랩의 왼쪽 [파일] 탭에 드래그하여 업로드한 후 아래 셀을 실행하세요.

```python
import os
import zipfile

zip_path = 'dataset.zip'
extract_path = '/content/'

if os.path.exists(zip_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print("압축 해제 완료!")
else:
    print("dataset.zip 파일을 먼저 업로드해주세요!")
```

### 4. data.yaml 경로 수정
코랩 환경에 맞춰서 `data.yaml`의 `path`를 수정해줍니다.

```python
import yaml

yaml_path = '/content/data/dataset/data.yaml'
with open(yaml_path, 'r') as f:
    config = yaml.safe_load(f)

# 코랩 내부 절대 경로로 수정
config['path'] = '/content/data/dataset'

with open(yaml_path, 'w') as f:
    yaml.dump(config, f)

print("data.yaml 경로 수정 완료!")
```

### 5. 학습 시작 (Fine-tuning)
이전에 결과가 좋았던 `best.pt` 모델(있다면 함께 업로드)을 기반으로 추가 학습을 하거나, 기본 모델(`yolov8n.pt`)로 시작합니다.

```python
from ultralytics import YOLO

# best.pt를 업로드했다면 'best.pt'로 설정, 처음부터라면 'yolov8n.pt'
model = YOLO('yolov8n.pt') 

model.train(
    data='/content/data/dataset/data.yaml', 
    epochs=100, 
    imgsz=640, 
    batch=16, 
    name='dashboard_ocr_finetune'
)
```
